```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
vsales_raw = pd.read_csv('https://raw.githubusercontent.com/paulaked/data25-visualisations/MattWoodburn/vgsales.csv')
vsales_raw
```

```python
vsales = vsales_raw.dropna()
vsales
```

```python
from matplotlib import rcParams

rcParams['figure.figsize'] = 15,8
sns.barplot(x="Platform", y="Global_Sales", data=vsales)
```

```python
from matplotlib import rcParams

rcParams['figure.figsize'] = 15,8
sns.lineplot(x="Year", y="NA_Sales", data=vsales)
```

```python
vsales_drop = vsales.drop(['Rank','Name','Platform', 'Genre','Publisher','Other_Sales', 'Global_Sales'], axis=1)
vsales_melt = vsales_drop.melt("Year", var_name="NA_Sales&EU_Sales", value_name = "Sales")
vsales_melt
```

```python
sns.lineplot(data=vsales_melt, x="Year", y="Sales", hue="NA_Sales&EU_Sales")
```
