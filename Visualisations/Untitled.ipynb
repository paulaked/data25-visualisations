```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
aggregation_functions = {'Name' : 'first', 'Year': 'first', 'Genre': 'first', 'Publisher': 'first',
                         'NA_Sales': 'sum', 'EU_Sales': 'sum', 'JP_Sales': 'sum',
                         'Other_Sales': 'sum', 'Global_Sales': 'sum'}
```

```python
aggregation_functions2 = {'Genre': 'first', 'NA_Sales': 'mean', 'EU_Sales': 'mean', 'JP_Sales': 'mean',
                         'Other_Sales': 'mean', 'Global_Sales': 'mean'}
```

```python
df
```

```python
df = df.iloc[:500,]
```

```python
df
```

```python
ea = df[df['Publisher'] == 'Electronic Arts'].reset_index(drop = True)
ea2 = ea.groupby('Name').aggregate(aggregation_functions)
ea2
```

```python
av = df[df['Publisher'] == 'Activision'].reset_index(drop = True)
av2 = av.groupby('Name').aggregate(aggregation_functions)
av2.groupby('Global_Sales').first()
```

```python
nin = df[df['Publisher'] == 'Nintendo'].reset_index(drop = True)
nin2 = nin.groupby('Name').aggregate(aggregation_functions)
nin2.groupby('Global_Sales').last()
```

```python
tt = df[df['Publisher'] == 'Take-Two Interactive'].reset_index(drop = True)
tt2 = tt.groupby('Name').aggregate(aggregation_functions)
tt2.groupby('Global_Sales').first()
```

```python
sy = df[df['Publisher'] == 'Sony Computer Entertainment'].reset_index(drop = True)
sy2 = sy.groupby('Name').aggregate(aggregation_functions)
sy2.groupby('Global_Sales').first()
```

```python
bs = df[df['Publisher'] == 'Bethesda Softworks'].reset_index(drop = True)
bs2 = bs.groupby('Name').aggregate(aggregation_functions)
bs2.groupby('Global_Sales').first()
```

```python
ub = df[df['Publisher'] == 'Ubisoft'].reset_index(drop = True)
ub2 = ub.groupby('Name').aggregate(aggregation_functions)
ub2
```

```python
sns.barplot(x='Name', y='Global_Sales', data = ub2)
```
