```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
apps = pd.read_csv('/Users/jenieka/Downloads/googleplaystore.csv')
appsdf = pd.DataFrame(apps)
appsdf 
#this is the original set of data put into a dataframe
```

```python
appsdf.drop_duplicates(inplace=True)
appsdf
#this is the "cleaned" dataframe 
```

```python
# del appsdf['Reviews']
# del appsdf['Current Ver']
# del appsdf['Android Ver']
# del appsdf['Last Updated']
appsdf
#Here I eliminated some columns that I felt weren't useful or relevant 
```

```python
appsdf_filtered = appsdf.dropna()
appsdf_filtered
#Again, cleaned data cus there was a lot of null values 
```

```python
paid_appsdf = appsdf_filtered[appsdf_filtered['Type'] == 'Paid']
sortedpaidapps = paid_appsdf.groupby('Price').mean()  
sortedpaidapps
#Here I selected out specifically the paid apps to see how (or if) this affects ratings
```

```python
sns.lineplot(x='Price', y='Rating', data = paid_appsdf)
plt.title('Avg Rating of Paid Apps vs Price') #but this isn't a very good graph at all 
#it's oook cus it does kiind of show the more expensive the lower the minimum rating is, but also the higher the
#max rating is 
```
