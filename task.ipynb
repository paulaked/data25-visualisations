```python
import pandas as pd
import numpy as np
```

```python
vid_games = pd.read_csv('/Users/alishaparmar/Downloads/vgsales.csv')
vid_games
```

```python
clean_data = vid_games[vid_games['Global_Sales']>=0.5]
clean_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x='Global_Sales', y='Genre', data = clean_data)
plt.title('The Global Sales for each Genre')
```

```python
sns.barplot(x='Global_Sales', y='Platform', data = clean_data.head(50))
plt.title('The Global Sales for each Platform')
```

```python
sns.lineplot(x='Year', y ='Global_Sales', data = clean_data)
plt.title('The Global Sales across the Years')
```

```python
data = clean_data.describe()
data
```

```python
x = clean_data.groupby('Publisher')['Global_Sales'].mean().sort_values().tail(15)
x.plot(kind ='barh')
```
